## Create and Display

```python
class node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
        
class linkedlist:
    def __init__(self):
        self.head = None
        
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new
        new.prev = temp
        
    def display(self):
        if self.head is None:
            print("head is none")
            return
        temp = self.head
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")

obj = linkedlist()

obj.add_value(10)
obj.add_value(20)
obj.add_value(30)
obj.add_value(40)

obj.display()
```

## Insertion in Begin, End and Given Position 

```python
class node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
        
class linkedlist:
    def __init__(self):
        self.head = None
    
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next 
        temp.next = new
        new.prev = temp
        
    def insert_pos(self, data, pos):
        new = node(data)
        if pos == 1:
            new.next = self.head
            self.head.prev = new
            self.head = new
            return 
        temp = self.head
        for i in range(pos - 2):
            temp = temp.next
        new.next = temp.next
        new.prev = temp
        if temp.next:
            temp.next.prev = new
        temp.next = new 
        
    def display(self):
        if self.head is None:
            print("head is a none")
            return
        temp = self.head 
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")
            
obj = linkedlist()

obj.add_value(10)
obj.add_value(20)
obj.add_value(30)

obj.display()

print("first position insertion")
obj.insert_pos(5, 1)
obj.display()

print("middle position insertion")
obj.insert_pos(15, 3)
obj.display()

print("end position insertion")
obj.insert_pos(40, 6)
obj.display()
```

## Reverse the Given List 

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class linkedlist:
    def __init__(self):
        self.head = None
    
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head 
        while temp.next:
            temp = temp.next
        temp.next = new
        new.prev = temp
        
    def reverse(self):
        temp = None
        curr = self.head
        while curr:
            temp = curr.prev
            curr.prev = curr.next
            curr.next = temp
            curr = curr.prev
        self.head = temp.prev
        
    def display(self):
        if self.head is None:
            print("head is none")
            return
        temp = self.head
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")
            
obj = linkedlist()

obj.add_value(10)
obj.add_value(20)
obj.add_value(30)
obj.add_value(40)

print("the original list is")
obj.display()

print("the reversed list is")
obj.reverse()
obj.display()
```
