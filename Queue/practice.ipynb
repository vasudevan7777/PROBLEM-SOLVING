## CREATE AND DISPLAY

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class queue:
    def __init__(self):
        self.front = None
        self.rear = None
        
    def enqueue(self, data):
        new = node(data)
        if self.front is None:
            self.front = new
            self.rear = new
            return
        self.rear.next = new
        self.rear = new
    
    def dequeue(self):
        if self.front is None:
            print("queue is empty")
            return
        temp = self.front
        self.front = self.front.next
        
        if self.front is None:
            self.rear = None
            
    def display(self):
        if self.front is None:
            print("queue is empty")
            return
        temp = self.front
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")

obj = queue()

obj.enqueue(10)
obj.enqueue(20)
obj.enqueue(30)
obj.enqueue(40)

obj.display()

obj.dequeue()
obj.display()
```

## REVERSE THE GIVEN LIST

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class queue:
    def __init__(self):
        self.front = None
        self.rear = None
    
    def enqueue(self, data):
        new = node(data)
        if self.front is None:
            self.front = new
            self.rear = new
            return
        self.rear.next = new
        self.rear = new
        
    def reverse(self):
        prev = None
        curr = self.front
        self.rear = self.front
        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
        self.front = prev
    
    def display(self):
        if self.front is None:
            print("queue is empty")
            return 
        temp = self.front
        while temp:
            print(temp.data, end = '->')
            temp = temp.next
        print("NULL")
        
obj = queue()

obj.enqueue(10)
obj.enqueue(20)
obj.enqueue(30)
obj.enqueue(40)

obj.display()

obj.reverse()
obj.display()
```
