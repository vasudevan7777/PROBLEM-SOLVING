## BASIC PROBLEMS 

```python
# print the number upto n values
def recursion(val, num):
    if val > num:
        return    
    print(val, end = " ") 
    recursion(val+1, num)  

num = 8
print("normal recursion")
recursion(0, num)   

print('\n')
#####################################################################          

# print the number upto n values in reverse
def recursion(num):
    if num < 0:
        return
    print(num, end = " ")
    recursion(num - 1)
 
num = 8
print("reverse recursion")
recursion(num)

print('\n')
#####################################################################

# recursion function using backtracking print 1 from 1 to N
def recursion(val, num):
    if val < 0:
        return 
    recursion(val-1, num)
    print(val, end = " ")

num = 10
print("recursion using simple backtracking")
recursion(num, num)

print('\n')
#####################################################################

# recursion function using backtracking print N from N to 1
def recursion(val, num):
    if val > num:
        return 
    recursion(val+1, num)
    print(val, end = " ")

num = 10
print("recursion using simple backtracking in reverse")
recursion(1, num)

print('\n')
#####################################################################

# sum upto n time of the given value 
def sumvalue(val, tot):
    if val < 0:
        print(tot)
        return 
    sumvalue(val-1, tot + val)
    
num = 5
print("sum upto n times of the given value")
sumvalue(num, 0)

print('\n')
#####################################################################

# factorial of a given number
def factorial(val):
    if val == 1:
        return val
    return val * factorial(val-1)

val = 4
print("the factorial of the given number")
print(factorial(val))

print('\n')
#####################################################################

# fabonacci of the given number
def fibonacci(num):
    if num <= 1:
        return num
    return fibonacci(num-1) + fibonacci(num-2)
    
num = 10
print("fabonacci of the given number")
print(fibonacci(num))

print('\n')
#####################################################################

# reverse the list
def reverse_recursion(lst, ft, lt):
    if ft > lt:
        print(lst)
        return
    lst[ft], lst[lt] = lst[lt], lst[ft]
    reverse_recursion(lst, ft+1, lt-1)
    
lst = [5, 4, 3, 2, 1, 0]
print("reverse the given list")
reverse_recursion(lst, 0, len(lst)-1)

print('\n')
#####################################################################

# check if a string is a palindrome or not
def recursion(st, ft, lt, check):
    if ft > lt:
        print(check)
        return
    if st[ft] == st[lt]:
        check = True
    else:
        check = False
    recursion(st, ft+1, lt-1, check)

st = "malayalam"
check = True
print("the given sring is a palindrome or not")
recursion(st, 0, len(st)-1, check)
```
