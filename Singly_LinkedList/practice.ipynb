## Create and Display

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class linkedlist:
    def __init__(self):
        self.head = None
        
    def add_begin(self, data):
        new = node(data)
        new.next = self.head
        self.head = new
    
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head 
        while temp.next:
            temp = temp.next
        temp.next = new
        
    def display(self):
        if self.head is None:
            print("head is none")
            return 
        temp = self.head
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")

obj = linkedlist()

obj.add_value(10)
obj.add_value(20)
obj.add_value(30)
obj.display()

obj.add_begin(5)
obj.display()
```

## Insertion in Begin, End, and Given Position

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class linkedlist:
    def __init__(self):
        self.head = None
    
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new
    
    def insert_pos(self, pos, data):
        if pos < 1:
            print("position out of range")
            return 
        new = node(data)
        if pos == 1:
            new.next = self.head
            self.head = new
            return
        temp = self.head
        for i in range(pos - 2):
            temp = temp.next
        new.next = temp.next
        temp.next = new
        
    def display(self):
        if self.head is None:
            print("head is none")
            return
        temp = self.head
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")

obj = linkedlist()
obj.add_value(10)
obj.add_value(20)
obj.add_value(30)
obj.display()

obj.insert_pos(1, 5)
obj.display()
obj.insert_pos(2, 15)
obj.display()
obj.insert_pos(5, 40)
obj.display()
```

## Deletion in Begin, End and Given Position

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class linkedlist:
    def __init__(self):
        self.head = None
    
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head 
        while temp.next:
            temp = temp.next
        temp.next = new

    def delete_pos(self, pos):
        if pos < 1:
            print("position out of range")
            return
        if pos == 1:
            temp = self.head
            self.head = temp.next
            temp.next = None
            return
        temp = self.head
        for i in range(pos - 2):
            temp = temp.next
        if temp.next is None:
            print("position out of range")
            return
        temp.next = temp.next.next
        
    def display(self):
        if self.head is None:
            print("head is none")
            return 
        temp = self.head
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")

obj = linkedlist()

obj.add_value(10)
obj.add_value(20)
obj.add_value(30)
obj.add_value(40)
obj.add_value(50)
obj.display()

obj.delete_pos(1)
obj.display()

obj.delete_pos(3)
obj.display()
```

## Reverse the Given List

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class linkedlist:
    def __init__(self):
        self.head = None
    
    def add_value(self, data):
        new = node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new
        
    def reverse_list(self):
        prev = None
        curr = self.head
        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
        self.head = prev
        
    def display(self):
        if self.head is None:
            print("head is none")
            return
        temp = self.head
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")
        
obj = linkedlist()

obj.add_value(10)
obj.add_value(20)
obj.add_value(30)
obj.add_value(40)

obj.display()

print("the reversed list are")
obj.reverse_list()
obj.display()
```
