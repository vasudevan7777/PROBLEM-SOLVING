## Middle of the linked list - LC 876

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow
```

## Linked list cycle - LC 141

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        slow = fast = head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False
```

## Remove Duplicates from Sorted List I - LC 83

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        lst = []
        temp = head
        prev = None
        while temp:
            if temp.val not in lst:
                lst.append(temp.val)
                prev = temp
            else:
                prev.next = temp.next
            temp = temp.next
        return head
```

## Merge Two Sorted Lists - LC 21

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        temp = dum = ListNode()
        while list1 and list2:
            if list1.val < list2.val:
                temp.next = list1
                list1 = list1.next
            else:
                temp.next = list2
                list2 = list2.next
            temp = temp.next
        temp.next = list1 if list1 else list2
        return dum.next
```

## Remove Linked List Elements - LC 203

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        if head is None:
            return 
        dum = ListNode()
        dum.next = head
        curr = head
        prev = dum
        while curr:
            if curr.val == val:
                prev.next = curr.next
            else:
                prev = curr
            curr = curr.next
        return dum.next
```

## Add Two Numbers - LC 02

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverse(self, head):
        prev = None
        curr = head

        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt

        return prev

    def addTwoNumbers(self, l1, l2):

        l1 = self.reverse(l1)
        l2 = self.reverse(l2)

        num1 = num2 = ""

        while l1:
            num1 += str(l1.val)
            l1 = l1.next

        while l2:
            num2 += str(l2.val)
            l2 = l2.next

        total = int(num1) + int(num2)

        digits = list(str(total))[::-1]

        dummy = ListNode(0)
        curr = dummy

        for d in digits:
            curr.next = ListNode(int(d))
            curr = curr.next

        return dummy.next
```

## Remove Nth Node From End of List - LC 19

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dum = ListNode(0, head)
        slow = dum
        fast = head
        for i in range(n):
            fast = fast.next
        while fast is not None:
            slow = slow.next
            fast = fast.next
        slow.next = slow.next.next
        return dum.next
```

## Merge k Sorted Lists - LC 23

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        arr = []
        for ch in lists:
            while ch:
                arr.append(ch.val)
                ch = ch.next
        arr.sort()

        dum = ListNode(0)
        curr = dum
        for num in arr:
            curr.next = ListNode(num)
            curr = curr.next
        return dum.next
```

## Remove Duplicates from Sorted List II - LC 82

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        lst = []
        temp = head
        while temp:
            lst.append(temp.val)
            temp = temp.next
        
        st = head
        from collections import Counter
        cout = Counter(lst)
        prev = None
        for i in lst:
            if cout[i] == 1:
                st.val = i
                prev = st
                st = st.next

        if prev is None:
            return None

        prev.next = None
        return head
```

## Reverse Linked List II - LC 92

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        dum = ListNode(0, head)
        l, curr = dum, head

        for i in range(left-1):
            l, curr = curr, curr.next
        
        prev = None
        for i in range(right-left+1):
            temNext = curr.next
            curr.next = prev
            prev, curr = curr, temNext

        l.next.next = curr
        l.next = prev
        return dum.next
```

## Linked List Cycle II - LC 142

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = fast = head
        while fast and fast.next:
            slow = slow.next 
            fast = fast.next.next
            if slow == fast:
                break
        else:
            return None
        slow = head
        while slow != fast:
            slow = slow.next
            fast = fast.next
        return slow
```
