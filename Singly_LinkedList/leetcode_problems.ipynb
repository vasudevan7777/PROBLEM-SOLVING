## Middle of the linked list - LC 876

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow
```

## Linked list cycle - LC 141

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        slow = fast = head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False
```

## Remove Duplicates from Sorted List I - LC 83

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        lst = []
        temp = head
        prev = None
        while temp:
            if temp.val not in lst:
                lst.append(temp.val)
                prev = temp
            else:
                prev.next = temp.next
            temp = temp.next
        return head
```

## Merge Two Sorted Lists - LC 21

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        temp = dum = ListNode()
        while list1 and list2:
            if list1.val < list2.val:
                temp.next = list1
                list1 = list1.next
            else:
                temp.next = list2
                list2 = list2.next
            temp = temp.next
        temp.next = list1 if list1 else list2
        return dum.next
```

## Remove Linked List Elements - LC 203

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        if head is None:
            return 
        dum = ListNode()
        dum.next = head
        curr = head
        prev = dum
        while curr:
            if curr.val == val:
                prev.next = curr.next
            else:
                prev = curr
            curr = curr.next
        return dum.next
```

## Add Two Numbers - LC 02

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverse(self, head):
        prev = None
        curr = head

        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt

        return prev

    def addTwoNumbers(self, l1, l2):

        l1 = self.reverse(l1)
        l2 = self.reverse(l2)

        num1 = num2 = ""

        while l1:
            num1 += str(l1.val)
            l1 = l1.next

        while l2:
            num2 += str(l2.val)
            l2 = l2.next

        total = int(num1) + int(num2)

        digits = list(str(total))[::-1]

        dummy = ListNode(0)
        curr = dummy

        for d in digits:
            curr.next = ListNode(int(d))
            curr = curr.next

        return dummy.next
```

## Remove Nth Node From End of List - LC 19

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dum = ListNode(0, head)
        slow = dum
        fast = head
        for i in range(n):
            fast = fast.next
        while fast is not None:
            slow = slow.next
            fast = fast.next
        slow.next = slow.next.next
        return dum.next
```

## Merge k Sorted Lists - LC 23

## Remove Duplicates from Sorted List II - LC 82

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        lst = []
        temp = head
        while temp:
            lst.append(temp.val)
            temp = temp.next
        
        st = head
        from collections import Counter
        cout = Counter(lst)
        prev = None
        for i in lst:
            if cout[i] == 1:
                st.val = i
                prev = st
                st = st.next

        if prev is None:
            return None

        prev.next = None
        return head
```

## Reverse Linked List II - LC 92

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        dum = ListNode(0, head)
        l, curr = dum, head

        for i in range(left-1):
            l, curr = curr, curr.next
        
        prev = None
        for i in range(right-left+1):
            temNext = curr.next
            curr.next = prev
            prev, curr = curr, temNext

        l.next.next = curr
        l.next = prev
        return dum.next
```

## Linked List Cycle II - LC 142

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = fast = head
        while fast and fast.next:
            slow = slow.next 
            fast = fast.next.next
            if slow == fast:
                break
        else:
            return None
        slow = head
        while slow != fast:
            slow = slow.next
            fast = fast.next
        return slow
```

## Sort List - LC 148

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        lst = []
        temp = head
        while temp:
            lst.append(temp.val)
            temp = temp.next
        lst.sort()
        
        st = head
        indx = 0
        while st:
            st.val = lst[indx]
            st = st.next
            indx += 1
        return head
```

## Intersection of Two Linked Lists - LC 160

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        def length(head):
            count = 0
            while head:
                count += 1
                head = head.next
            return count
        
        lenA = length(headA)
        lenB = length(headB)
        p1 = headA
        p2 = headB
        if lenA > lenB:
            for _ in range(lenA - lenB):
                p1 = p1.next
        else:
            for _ in range(lenB - lenA):
                p2 = p2.next
        while p1 and p2:
            if p1 == p2:
                return p1
            p1 = p1.next
            p2 = p2.next
        return None
```

## Palindrome Linked List - LC 234

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return True

        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        prev = None
        while slow:
            nxt = slow.next
            slow.next = prev
            prev = slow
            slow = nxt

        left = head
        right = prev
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next

        return True
```

## Merge In Between Linked Lists - LC 1669

```python
class Solution:
    def mergeInBetween(self, list1: ListNode, a: int, b: int, list2: ListNode) -> ListNode:
        prev = list1
        for _ in range(a - 1):
            prev = prev.next

        curr = prev
        for _ in range(b - a + 2):
            curr = curr.next

        prev.next = list2

        tail = list2
        while tail.next:
            tail = tail.next

        tail.next = curr

        return list1
```

## Delete the Middle Node of a Linked List - LC 2095

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        temp = head 
        if temp.next is None:
            dum = ListNode()
            head.next = dum
            return dum.next
        if temp.next.next is None:
            temp.next = None
            return head
        slow = fast = head
        while fast and fast.next:
            prev = slow
            slow = slow.next
            temp = slow
            fast = fast.next.next
        prev.next = temp.next 
        temp.next = None
        temp = prev.next
        return head
```

## Merge Nodes in Between Zeros - LC 2181

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeNodes(self, head: Optional[ListNode]) -> Optional[ListNode]:
        curr = prev = head
        temp = head.next
        tot = check = 0
        while temp:
            if temp.val != 0:
                tot += temp.val
            else:
                if temp.val == 0:
                    if check == 0:
                        prev.val = tot
                        check = 1
                    else:
                        prev.next.val = tot
                        prev = prev.next
                    tot = 0
            temp = temp.next 
        prev.next = None
        return curr
```

## Reorder List - LC 143

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reorderList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: None Do not return anything, modify head in-place instead.
        """
        
        if not head or not head.next:
            return

        slow, fast = head, head
        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next

        prev, curr = None, slow.next
        slow.next = None 
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt

        ft, st = head, prev
        while st:
            tmp1 = ft.next
            tmp2 = st.next
            ft.next = st
            st.next = tmp1
            ft = tmp1
            st = tmp2
```

## Insertion Sort List - LC 147

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def insertionSortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        lst = []
        sap = head
        while sap:
            lst.append(sap.val)
            sap = sap.next 
        lst.sort()
        
        temp = dum = ListNode(0)
        for i in lst:
            temp.next = ListNode(i)
            temp = temp.next
        return dum.next
```
