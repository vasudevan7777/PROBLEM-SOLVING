## Create and Display

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class stack:
    def __init__(self):
        self.top = None
    
    def push(self, data):
        new = node(data)
        if self.top is None:
            self.top = new
            print(f"{data} is push into the stack")
            return
        new.next = self.top
        self.top = new
        print(f"{data} is push into the stack")
        
    def pop(self):
        if self.top is None:
            print("stack is empty")
            return
        temp = self.top
        self.top = self.top.next
        print(f"{temp.data} is pop from the stack")
        
    def peek(self):
        if self.top is None:
            print("stack is empty")
            return
        else:
            print(f"the top of the value in a stack is {self.top.data}")
            
    def display(self):
        if self.top is None:
            print("stack is empty")
            return
        temp = self.top
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")
            
obj = stack()

obj.push(10)
obj.push(20)
obj.push(30)
obj.push(40)

obj.display()

obj.peek()

obj.pop()
obj.display()

obj.peek()
```

## Middle of the Stack

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class stack:
    def __init__(self):
        self.top = None
    
    def push(self, data):
        new = node(data)
        if self.top is None:
            self.top = new
            print(f"{data} is push into the stack")
            return
        new.next = self.top
        self.top = new
        print(f"{data} is push into the stack")
    
    def pop(self):
        if self.top is None:
            print("stack is empty")
            return 
        temp = self.top
        self.top = self.top.next
        print(f"{temp.data} is pop from the stack")
        
    def middle(self):
        slow = fast = self.top
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        print(f"the middle value of the given stack is {slow.data}")
        
    def display(self):
        if self.top is None:
            print("stack is empty")
            return
        temp = self.top
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")
            
obj = stack()

obj.push(10)
obj.push(20)
obj.push(30)
obj.push(40)

obj.display()

obj.middle()
```

## Reverse the Stack

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class stack:
    def __init__(self):
        self.top = None
    
    def push(self, data):
        new = node(data)
        if self.top is None:
            self.top = new
            print(f"{data} is push into the stack")
            return
        new.next = self.top
        self.top = new
        print(f"{data} is push into the stack")
    
    def pop(self):
        if self.top is None:
            print("stack is empty")
            return 
        temp = self.top
        self.top = self.top.next
        print(f"{temp.data} is pop from the stack")
        
    def reverse(self):
        prev = None
        curr = self.top
        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr =  temp
        self.top = prev
        
    def display(self):
        if self.top is None:
            print("stack is empty")
            return
        temp = self.top
        while temp:
            print(temp.data, end = "->")
            temp = temp.next
        print("NULL")
            
obj = stack()

obj.push(10)
obj.push(20)
obj.push(30)
obj.push(40)

print('the original list is')
obj.display()

print('the reverse list is')
obj.reverse()
obj.display()
```
